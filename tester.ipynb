```python
import json

def compare_json_structures(obj1, obj2, path="root"):
    if type(obj1) != type(obj2):
        print(f"❌ Type mismatch at {path}: {type(obj1).__name__} vs {type(obj2).__name__}")
        return

    if isinstance(obj1, dict):
        keys1, keys2 = set(obj1.keys()), set(obj2.keys())

        for k in sorted(keys1 - keys2):
            print(f"❌ Missing in second: {path}.{k}")
        for k in sorted(keys2 - keys1):
            print(f"❌ Missing in first: {path}.{k}")

        for k in sorted(keys1 & keys2):
            compare_json_structures(obj1[k], obj2[k], f"{path}.{k}")

    elif isinstance(obj1, list):
        if len(obj1) != len(obj2):
            print(f"⚠️ Length mismatch at {path}: {len(obj1)} vs {len(obj2)}")
        for i, (a, b) in enumerate(zip(obj1, obj2)):
            compare_json_structures(a, b, f"{path}[{i}]")

    else:
        # Compare type-level (not value-level) differences
        if type(obj1) != type(obj2):
            print(f"❌ Value type mismatch at {path}: {type(obj1).__name__} vs {type(obj2).__name__}")

# --- Load your two topologies ---
with open("reg_topology.json") as f1, open("waxman_topology.json") as f2:
    topo_good = json.load(f1)
    topo_bad = json.load(f2)

# --- Run comparison ---
compare_json_structures(topo_good, topo_bad)
```
